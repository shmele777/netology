```python
# Задание 1
documents = [
    {'type': 'passport', 'number': '2207 876234', 'name': 'Василий Гупкин'},
    {'type': 'invoice', 'number': '11-2', 'name': 'Геннадий Покемонов'},
    {'type': 'insurance', 'number': '10006', 'name': 'Аристарх Павлов'}
]

directories = {
    '1': ['2207 876234', '11-2'],
    '2': ['10006'],
    '3': []
}


def owner_document(number_document):
    """
    Поиск документа в базе по его номеру и вывод имя владельца.
    
    number_document(string): номер документа.
    """

    for doc in documents:
        if doc['number'] == number_document:
            return 'Владелец документа: ' + doc['name']
    return 'Документ не найден в базе'


def num_directory(number_document):
    """
    Поиск документа в базе по его номеру и вывод номер полки.
    

    number_document(string): номер документа.
    """ 

    for n, docs in directories.items():
        if number_document in docs:
            return 'Документ хранится на полке: ' + n
    return 'Документ не найден в базе'


def all_infotion():
    """ Вывод полной информации по всем документам"""

    result = ''
    for doc in documents:
        result += f'№: {doc["number"]}, тип: {doc["type"]}, владелец: {doc["name"]}, полка хранения: {num_directory(doc["number"])}\n'
    return result


def current_directories():
    """ Выводит в строку все существующие полки"""

    return ", ".join(map(str, directories.keys()))


def add_directory(number_directory):
    """
    Добавляет новую полку, если такой ранее небыло.

    number_direcrory(string): номер полки, которую необходимо добавить.
    """

    result = ''
    if number_directory in directories.keys():
        result = 'Такая полка уже существует.'
    else:
        directories[number_directory] = []
        result = 'Полка добавлена.'
    return result + f' Текущий перечень полок: {current_directories()}.'


def del_directory(number_directory):
    """
    Удаляет существующую полку.

    number_direcrory(string): номер полки, которую необходимо удалить.
    """

    result = ''
    if number_directory in directories.keys():
        if directories[number_directory] == []:
            directories.pop(number_directory)
            result = 'Полка удалена.'
        else:
            result = 'На полке есть документа, удалите их перед удалением полки.'
    else:
        result = 'Такой полки не существует.'
    return result + f' Текущий перечень полок: {current_directories()}.'


def help_command():
    """ Выводит справочную информацию"""

    return '''
        p\t- имя ползователя по номеру документа
        s\t- номер полки по номеру документа
        l\t- полная информация по всем документам
        as\t- добавить новую полку
        ds\t- удалить полку, если она пуста
        h\t- справка
        q\t- завершение работы'
    '''



def main():
    while True:
        command = input('Введите команду (h - справка): ')
        
        if command in ('p', 's'):
            number_document = input('Введите номер документа: ')
            if command == 'p':
                print(owner_document(number_document))
            else:
                print(num_directory(number_document))
        elif command == 'l':
            print(all_infotion())
        elif command in ('as', 'ds'):
            number_directory = input('Введите номер полки: ')
            if command == 'as':
                print(add_directory(number_directory))
            else:
                print(del_directory(number_directory))
        elif command == 'h':
            print(help_command())
        elif command == 'q':
            break
        else:
            print(f'{command} - нет такой команды')

main()
```
