```python
import numpy as np

trials = 10000000   # Количество попыток

winning = {
    '777': {'win': 200, 'count': 0, 'chance': 0 },
    '999': {'win': 100, 'count': 0, 'chance': 0 },
    '555': {'win': 50, 'count': 0, 'chance': 0 },
    '333': {'win': 15, 'count': 0, 'chance': 0 },
    '111': {'win': 10, 'count': 0, 'chance': 0 },
    '*77': {'win': 5, 'count': 0, 'chance': 0 },
    '**7': {'win': 3, 'count': 0, 'chance': 0 },
    '*00': {'win': 2, 'count': 0, 'chance': 0 },
    '**0': {'win': 1, 'count': 0, 'chance': 0 }
} # win - сумма выйгрыша, count - количество совпадений, chance - вероятность 

for i in range(0, 1000):
    if i == 777:
        winning['777']['chance'] = round(winning['777']['chance'] + 0.001, 3)
        continue

    elif i == 999:
        winning['999']['chance'] = round(winning['999']['chance'] + 0.001, 3)
        continue

    elif i == 555:
        winning['555']['chance'] = round(winning['555']['chance'] + 0.001, 3)
        continue

    elif i == 333:
        winning['333']['chance'] = round(winning['333']['chance'] + 0.001, 3)
        continue

    elif i == 111:
        winning['111']['chance'] = round(winning['111']['chance'] + 0.001, 3)
        continue

    elif i%100 == 77:
        winning['*77']['chance'] = round(winning['*77']['chance'] + 0.001, 3)
        continue

    elif i%10 == 7:
        winning['**7']['chance'] = round(winning['**7']['chance'] + 0.001, 3)
        continue

    elif i%100 == 0:
        winning['*00']['chance'] = round(winning['*00']['chance'] + 0.001, 3)
        continue

    elif i%10 == 0:
        winning['**0']['chance'] = round(winning['**0']['chance'] + 0.001, 3)
        continue

for i in range(trials):
    random_score = np.random.randint(0, 1000)

    if random_score == 777:
        winning['777']['count'] += 1
        continue

    elif random_score == 999:
        winning['999']['count'] += 1
        continue

    elif random_score == 555:
        winning['555']['count'] += 1
        continue

    elif random_score == 333:
        winning['333']['count'] += 1
        continue

    elif random_score == 111:
        winning['111']['count'] += 1
        continue

    elif random_score%100 == 77:
        winning['*77']['count'] += 1
        continue

    elif random_score%10 == 7:
        winning['**7']['count'] += 1
        continue

    elif random_score%100 == 0:
        winning['*00']['count'] += 1
        continue

    elif random_score%10 == 0:
        winning['**0']['count'] += 1
        continue

successes_all =  sum([x['count'] for x in winning.values()])
print(f'Вероятность хоть какого-то выйгрыша: {(successes_all/trials):.2%}')

result_sum = 0
result_sum_1 = 0
for x in winning.values():
    result_sum += x['win']*x['count']
    result_sum_1 += x['win']*x['chance']


print(f'Стредняя сумма выйгрыша, смоделированная: {result_sum/trials}')
    
print(f'Стредняя сумма выйгрыша, теоритическая: {result_sum_1}')
```
